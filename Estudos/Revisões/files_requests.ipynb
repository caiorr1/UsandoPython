```python
import os
import json
import requests

import glob as gl
```

```python
# Verifica se existe uma pasta com o nome "arquivos"

if not os.path.exists("arquivos"):
    
    # Caso não exista a pasta "arquivos" esse comando cria a pasta
    os.makedirs("arquivos")
```

```python
# Cria um arquivo TXT no diretório "arquivos"

with open(fr'.\arquivos\arquivo.txt', 'w') as arquivo:
    arquivo.write(f'Conteúdo do arquivo!!!')
```

```python
# Lê um arquivo TXT do diretório "arquivos" e exibe o conteúdo na tela

with open(fr'.\arquivos\arquivo.txt') as arquivo:
    print(arquivo.read())
```

```python
# Cria um dicionário

dicionario_json = {
    'nome': 'Vagner',
    'idade': 37,
    'altura': 1.70,
    'usuario_python': True
}

# Salva o dicionário como JSON

with open(fr'.\arquivos\dicionario.json', 'w') as arquivo:
    json.dump(dicionario_json, arquivo)
```

```python
# Lê o JSON como dicionário e exibe o conteúdo de uma chave na tela

with open(fr'.\arquivos\dicionario.json') as arquivo:
    dicionario_carregado = json.load(arquivo)
    print(dicionario_carregado['usuario_python'])
```

```python
for i in range(10):

    if gl.glob(fr'.\arquivos\arquivo_{i}.txt'):
        
        print(f'Arquivo {i} já existe!')
        
    else:
        print(f'Salvando arquivo {i}!')
        
        with open(fr'.\arquivos\arquivo_{i}.txt', 'x') as arquivo:
            
            for j in range(5):
                arquivo.writelines(f'Conteúdo do arquivo {i}, na linha {j}\n')
```
